```python
class CRNNGAN:
    def generator(self,shape):
        model = Sequential()
        model.add(LSTM(128,inpt_shape=shape,return_sequences=True))
        model.add(LSTM(128,inpt_shape=shape,return_sequences=True))
        model.add(Dense(32,activation="softmax"))
        return model
        
    def discriminator(self,shape):
        model = Sequential()
        model.add(Bidirectional(LSTM(128,input_shape=shape,return_sequences=True)))
        model.add(Bidirectional(LSTM(128,input_shape=shape,return_sequences=True)))
        model.add(Dense())
        
```

```python
import os
import midi,numpy
def generator_data(max_len):
    while True:
        artists = os.listdir("data/classical")
        for artist in artists:
            midis = os.listdir("data/classical/"+artist)
            for filename in midis:
                print("reading ","data/classical/{}/{}".format(artist,filename))
                pattern = midi.read_midifile("data/classical/{}/{}".format(artist,filename))
                print(len(pattern))
                if len(pattern) <= 2:
                    print("length less than 2")
                    continue
                tones = []
                velocities = []
                ticks = []
                length = 0
                for track in pattern[1:]:
                    if len(track) < max_len + 5:
                        continue
                    try:
                        tone = [float(x.data[0]) for x in track[5:-1]]
                        velocity = [float(x.data[1]) for x in track[5:-1]]
                        tick = [float(x.tick) for x in track[5:-1]]
                        tones.append(tone)
                        velocities.append(velocity)
                        ticks.append(tick)
                    except:
                        continue
                if len(tones) < 2:
                    continue
                length = min(len(tones[0]),len(tones[1]))
                print("length ",length)
                if length < max_len:
                    continue
                for i in range(2):
                    tones[i] = tones[i][:length]
                    ticks[i] = ticks[i][:length]
                    velocities[i] = velocities[i][:length]
                tones = tones[:2]
                velocities = velocities[:2]
                ticks = ticks[:2]
                X = []
                for i in range(max_len):
                    X += [[tones[0][i],tones[1][i],velocities[0][i],velocities[1][i],ticks[0][i],ticks[1][i]]]
                Y = [tones[0][max_len],tones[1][max_len],velocities[0][max_len],velocities[1][max_len],ticks[0][max_len],ticks[1][max_len]]
                yield numpy.array(X).reshape(max_len,6,1),numpy.array(Y)
                for i in range(max_len+1,length):
                    X += [Y]
                    X = X[1:]
                    Y = [tones[0][i],tones[1][i],velocities[0][i],velocities[1][i],ticks[0][i],ticks[1][i]]
#                     X_train = 
                    yield numpy.array(X).reshape(max_len,6,1),numpy.array(Y)
                
                
```

```python
i = 0
try:
    for data in generator_data(5):
        i += 1
        print(data)
        if i == 5:
            break
except Exception as e:
    print("err",e)#t
    
```

```python
import keras
max_len = 5
step_size = 10
def simpleRNN(size):
        model = keras.models.Sequential()
        model.add(keras.layers.LSTM(32,input_shape=size,return_sequences=True))
#         model.add(keras.layers.LSTM(32,input_shape=size,return_sequences=True))
        model.add(keras.layers.LSTM(size[1],return_sequences=True))
        model.add(keras.layers.Dense(size[1],activation="relu"))
        return model
# simpleRNN((max_len,6))
```

```python
model  = simpleRNN((max_len+1,6))
model.compile(loss='mean_absolute_error', optimizer="adam")
model.fit_generator(generator_data(max_len),steps_per_epoch=10)
# model.summary()
```
